**ASSIGNMENT 1 - SEARCH**

Below are codes providing basic implementations of BFS, DFS< and A* search/pathfinding algorithm. With tihs code, we have adapted everything from the graph representation, node coordinates, and heuristic functions to the specific input data and requirements provided by the assignment given.

Brief descriptions of each algorithm are as follows:



*   **Breadth-First Search (BFS)**: This algorithm explores all of the nodes at the current level before moving to the next level in a graph. This starts from the initial node and systematically explores all of the neighbors before moving onto their neighbors. BFS ensures that the first path is found from the start node tot he goal node is the shortest path. In the provided code, we have used a queue in order to keep track of the ndoes to be explored. We begin with the start node and continue adding the neighbors to the queue, exploring each level of nodes before moving deeper. THen we take the visited set, and this helps us make sure we don't revisit what we have already explored in the path.
*   **Depth-First Search (DFS)**: DFS explores as far down a branch as it possibly can before it begins to backtrack. This traverses deep into the graph before exploring siblings of the current node. DFS doesn't guarantee the shortest path unfortunately. Instead of using queue we use a stack to keep track of nodes being explored. We use the start node and the visited set once again. Then when we discover the goal, we return to the path from the beginning to the goal.
*   **A* Search**: A* is an informed search unlike BFS and DFS which uses a heuristic to guide the search toward the most promising nodes. THis combines the cost to reach the node (or otherwise known as the path cost) and then an estimate of the remaining cost to reach the goal or heuristic. A* guarantees the shortest path given the heuristic is admissble, which means to never overestimate the true cost. IN this algorithm, we use a priority queue or heapt to prioritize nodes based around a combination of the path cost and heuristic estimate. Then the heuristic function estimates the cost from the current node to the goal node. Manhattan distance and Euclidean distance are common heuristics but we utilize Euclidean here. Nodes are expanded based on their combined priority--that means that considers the path cost and heuristic estimate as aforementioned. Finally when we reach the goal node, we return to the beginning to goal.





```python
import heapq  # For A* algorithm

# graph
graph = {}

# node coordinates
node_coordinates = {}

# Define the test case file names
test_case_files = ["TestCase_01_EdgeList.txt", "TestCase_02_EdgeList.txt",
                   "TestCase_03_EdgeList.txt"]

# Iterate over the test cases
for case_number, test_case_file in enumerate(test_case_files, start=1):
    # Read the edge list from the file and update the graph
    with open(test_case_file, "r") as edge_list_file:
        for line in edge_list_file:
            n1, n2, w = line.strip().split(',')
            w = float(w)
            if n1 not in graph:
                graph[n1] = [(n2, w)]
            else:
                graph[n1].append((n2, w))
            if n2 not in graph:
                graph[n2] = [(n1, w)]
            else:
                graph[n2].append((n1, w))

    # Read node coordinates for the current test case
    node_coordinates_file = f"TestCase_{str(3).zfill(2)}_NodeID.csv"
    with open(node_coordinates_file, "r") as node_coordinates_file:
        for line in node_coordinates_file:
            n, x, y = line.strip().split(',')
            x, y = float(x), float(y)
            node_coordinates[n] = (x, y)

    # Print the graph and node coordinates for the current test case
    print(f"Test Case {3}:")
    print("Graph:", graph)
    print("Node Coordinates:", node_coordinates)

    # ---------------------
    # Breadth-First Search
    # ---------------------

    def bfs(graph, start, goal):
        visited = set()
        queue = [[start]]

        while queue:
            path = queue.pop(0)
            node = path[-1]

            if node == goal:
                return path

            if node not in visited:
                for neighbor, _ in graph[node]:
                    new_path = list(path)
                    new_path.append(neighbor)
                    queue.append(new_path)

                visited.add(node)

        return None

    # Example usage for BFS
    start_node = 'N_0'
    goal_node = 'N_24'
    bfs_path = bfs(graph, start_node, goal_node)
    print("BFS:", bfs_path)

    # ---------------------
    # Depth-First Search
    # ---------------------

    def dfs(graph, start, goal):
        visited = set()
        stack = [[start]]

        while stack:
            path = stack.pop()
            node = path[-1]

            if node == goal:
                return path

            if node not in visited:
                for neighbor, _ in graph[node]:
                    new_path = list(path)
                    new_path.append(neighbor)
                    stack.append(new_path)

                visited.add(node)

        return None

    # Example usage for DFS
    dfs_path = dfs(graph, start_node, goal_node)
    print("DFS:", dfs_path)

    # ----------------
    # A* Search
    # ----------------

    def heuristic(node, goal_node, node_coordinates):
        # Example heuristic: Euclidean distance
        x1, y1 = node_coordinates[node]
        x2, y2 = node_coordinates[goal_node]
        return ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5

    def astar(graph, start, goal, node_coordinates):
        open_list = [(0, start, [])]
        closed_set = set()

        while open_list:
            _, node, path = heapq.heappop(open_list)

            if node == goal:
                return path + [node]

            if node in closed_set:
                continue

            closed_set.add(node)

            for neighbor, cost in graph[node]:
                if neighbor not in closed_set:
                    priority = len(path) + cost + heuristic(neighbor, goal,
                                                            node_coordinates)
                    heapq.heappush(open_list, (priority, neighbor, path +
                     [node]))

        return None

    # Example usage for A*
    astar_path = astar(graph, start_node, goal_node, node_coordinates)
    print("A*:", astar_path)
    print()
```
